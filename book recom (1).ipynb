Collaborating filtering based recommender system

```python
ratings_with_name.groupby('User-ID').count()['Book-Rating'] 
```

```python
#upar se hme ye pta chalta hai ki hmare pass only 83643 hi ese log hai 
#jinhone aaj tak books ko rating di hai mtlb itne huge dataset set me only kuch hi hai padhe likhe
```

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating'] >200
x[x]
```

```python
#only 737 hi users hai jinhone aaj tak 200 se zyada books par rating di hai 
```

```python
x = ratings_with_name.groupby('User-ID').count()['Book-Rating'] > 200
padhe_likhe_users = x[x].index
```

```python
filtered_rating = ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
#ye wo log hai jinhone ratings kri hai 200 se zyada
```

```python
y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books = y[y].index
#books based filtering jis book par 50 kam se kam ratings hui hai 
```

```python
final_ratings = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_ratings
```

```python
#hmare pass finally 48546 rows bache hai ab bss pivot tables tgane hai.
```

```python
final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
#ab hmare pss wo hai users jisne 200 se zyada books par rating di hai and 50 se zyada books par ratings hui hai 
```

```python
pt = final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
pt
```

```python
#har book 1 vector hai in 735 dimensions
#to hmm coordinate krenge 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(pt)
```

```python
similarity_scores.shape
```

```python
def recommend(book_name):
    # index fetch
    index = np.where(pt.index==book_name)[0][0]
    similar_items = sorted(list(enumerate(similarity_scores[index])),key=lambda x:x[1],reverse=True)[1:5]
    
    data = []
    for i in similar_items:
        item = []
        temp_df = books[books['Book-Title'] == pt.index[i[0]]]
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Title'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Book-Author'].values))
        item.extend(list(temp_df.drop_duplicates('Book-Title')['Image-URL-M'].values))
        
        data.append(item)
    
    return data
#extend use kar rhe hai bcz agar append kreng eto 2D ban jayega
```

```python
recommend('1984')
```
